```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\ASUS\Downloads\Euro_2024_Matches.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['attendance'] = df['attendance'].str.replace(',', '')
df.head()
```

```python
df['attendance'] = df['attendance'].astype(int)
df.info()
```

```python
plt.figure(figsize=(16, 9))
plt.title('Attendance by Stadium')
plt.xticks(rotation=45)
sns.barplot(data=df, x='stadium', y ='attendance', errorbar=None, palette='pastel')
plt.show()
```

```python
sns.catplot(data=df, x='home_team', y='attendance',kind='bar', hue='stadium', errorbar=None, height=5,aspect=1.5)
plt.xticks(rotation=45)
plt.xlabel('Home Team')
plt.ylabel('Attendance')
plt.title('Attendance for home teams by stadium')
plt.show()
```

```python
sns.catplot(data=df, x='away_team', y='attendance',kind='bar', hue='stadium', errorbar=None, height=5,aspect=1.5)
plt.xticks(rotation=45)
plt.xlabel('Away Team')
plt.ylabel('Attendance')
plt.title('Attendance for away teams by stadium')
plt.show()
```

```python
df.drop(columns=['Home Total shots.', 'Home Shots on target.'], inplace=True)
df.head()
```

```python
df['Home Shots Ratio'] = df['Home Shots on target'] / df['Home Shots off target']
df.head()
```

```python
df['Away Shots Ratio'] = df['Away Shots on target'] / df['Away Shots off target']
df.head()
```

```python
plt.figure(figsize=(16,9))
sns.barplot(data=df, x='home_team', y='Home Shots Ratio', errorbar=None)
plt.xticks(rotation=45)
plt.xlabel('Home Team')
plt.title('Shots Ratio for Home Teams')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.barplot(data=df, x='away_team', y='Away Shots Ratio', errorbar=None)
plt.xticks(rotation=45)
plt.xlabel('Away Team')
plt.title('Shots Ratio for Away Teams')
plt.show()
```
